This cell reads an image from the specified file path using OpenCV and prints its pixel data (represented as a NumPy array).

```python
import cv2

path = '/gdrive/MyDrive/images/cats/h.jpeg'
im = cv2.imread(path)
print(im)
```

This cell displays the loaded image using `cv2_imshow` from `google.colab.patches`, which is necessary for displaying OpenCV images in Google Colab.

```python
from google.colab.patches import cv2_imshow
cv2_imshow(im)
```

This cell rotates the loaded image by 180 degrees using OpenCV's `rotate` function and then displays the rotated image.

```python
from google.colab.patches import cv2_imshow
rotated = cv2.rotate(im, cv2.ROTATE_180)
cv2_imshow(rotated)
```

This cell applies two different blurring filters to the image using OpenCV: Gaussian blur and Bilateral filter, and then displays the results of each.

```python
from google.colab.patches import cv2_imshow
Gaussian = cv2.GaussianBlur(im, (3, 3), 1)
cv2_imshow(Gaussian)

bilateral = cv2.bilateralFilter(im, 15, 150, 150)
cv2_imshow(bilateral)
```

This cell detects edges in the loaded image using the Canny edge detection algorithm from OpenCV and displays the resulting edge map.

```python
from google.colab.patches import cv2_imshow
edges = cv2.Canny(im, 100, 500)
cv2_imshow(edges)
```

This cell converts the loaded color image to grayscale using OpenCV's `cvtColor` function and displays the grayscale image.

```python
from google.colab.patches import cv2_imshow
gray_image = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

This cell creates a 3x3 NumPy array filled with ones, which is often used as a kernel in image processing operations.

```python
import numpy as np
kernel = np.ones((3, 3), np.uint8)
kernel
```

This cell dilates the image using the defined kernel and displays the resulting dilated image.

```python
from google.colab.patches import cv2_imshow
import numpy as np

kernel = np.ones((3, 3), np.uint8)
dilated = cv2.dilate(im, kernel, iterations=2)
cv2_imshow(dilated)
```

This cell applies a median blur filter to the image using OpenCV and displays the blurred image.

```python
from google.colab.patches import cv2_imshow
median = cv2.medianBlur(im, 11)
cv2_imshow(median)
```

This cell displays the loaded image using Matplotlib's `imshow` function with a 'gray' colormap and then displays the image again using `cv2_imshow`.

```python
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
plt.imshow(im, cmap='gray')
plt.show()
cv2_imshow(im)
```

This cell resizes the loaded image to a width and height of 100 pixels using OpenCV's `resize` function and displays the resized image.

```python
from google.colab.patches import cv2_imshow
resized = cv2.resize(im, (100, 100))
cv2_imshow(resized)
```
